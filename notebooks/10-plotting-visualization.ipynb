```df.plot(subplots=True, title='TITLE')``` ‚Äì Plots each numeric column in a separate subplot with a title.  
```df.plot(kind='line', title='TITLE')``` ‚Äì Plots all columns as line charts with a title.  
```df[['COLUMN_1', 'COLUMN_2']].plot(kind='line', title='TITLE')``` ‚Äì Plots two columns as separate lines on the same plot.  
```df.plot(kind='scatter', x='COLUMN_X', y='COLUMN_Y', title='TITLE')``` ‚Äì Creates a scatter plot between two variables.  
```df.plot(kind='hexbin', x='COLUMN_X', y='COLUMN_Y', title='TITLE')``` ‚Äì Creates a hexbin density plot for two variables.  
```df[['COLUMN_1', 'COLUMN_2']].plot(stacked=True, title='TITLE')``` ‚Äì Plots multiple columns as a stacked area chart.  
```df['COLUMN'].plot(kind='line', title='TITLE')``` ‚Äì Creates a line chart of a single column.  
```df['COLUMN'].plot(kind='area', title='TITLE')``` ‚Äì Creates an area chart of a single column.  
```df['COLUMN'].plot(kind='bar', title='TITLE')``` ‚Äì Creates a bar chart of a single column.  
```df['COLUMN'].plot(kind='hist', bins=N, title='TITLE')``` ‚Äì Creates a histogram with N bins.  
```df['COLUMN'].plot(kind='box', title='TITLE')``` ‚Äì Creates a box plot for a single column.  
```df['COLUMN'].value_counts().plot(kind='pie', title='TITLE')``` ‚Äì Creates a pie chart from value counts of a column.

````python
df.plot(subplots=True, title='Weather conditions')
````

### üé® Basic Customizations

```df[['COLUMN_1', 'COLUMN_2']].plot(kind='KIND', title='TITLE', xlabel='LABEL', ylabel='LABEL', color=['COLOR_1', 'COLOR_2'], figsize=(WIDTH, HEIGHT), grid=True, legend=True, alpha=VALUE)``` ‚Äì Creates a customized line chart with labels, size, grid, colors, and transparency.


````python
jan_df[['temp_max', 'temp_min']].plot(
    kind='line', 
    title='Change in Temperature over Time', 
    xlabel='Time', 
    ylabel='Temperature', 
    color=['red', 'blue'],
    figsize=(12, 6),
    grid=True,
    legend=True,
    alpha=0.4
)
````

### üîß Plot Configuration

```df['COLUMN'].plot(title='TITLE', figsize=(WIDTH, HEIGHT), style='STYLE', color='COLOR', linewidth=VALUE, alpha=VALUE, xlim=('START', 'END'), ylim=(LOW, HIGH), rot=ANGLE)``` ‚Äì Creates a styled line plot with axis limits, rotation, and transparency.  
```axis2 = df['COLUMN_2'].plot(ax=axis1, secondary_y=True, ...)``` ‚Äì Overlays a second line with a secondary y-axis.   
```axis2.set_ylabel('LABEL')``` ‚Äì Sets y-axis label for the secondary axis.   
```axis1.set_ylabel('LABEL')``` ‚Äì Sets y-axis label for the primary axis.   
```axis1.legend(['LABEL'], loc='LOCATION')``` ‚Äì Adds a legend for the first plot.   
```axis1.right_ax.legend(['LABEL'], loc='LOCATION')``` ‚Äì Adds a legend for the second axis.   
```plt.show()``` ‚Äì Displays the plot.  

````python
max_temp_axis = df['temp_max'].plot(
    title='Max Temperature',
    figsize=(12, 6),
    style='--o',
    color='red',
    linewidth=2,
    alpha=0.8,                 
    xlim=('2012-01', '2012-03'),
    ylim=(0, 25),
    rot=45,                  
)
max_temp_axis
````

````python
max_temp_axis = df['temp_max'].plot(
    title='Max Temperature with Wind Speed Overlay',
    figsize=(12, 6),
    style='--o',
    color='red',
    linewidth=2,
    alpha=0.8,                 
    xlim=('2012-01', '2012-03'),
    ylim=(0, 25),   
    rot=45
)

wind_axis = df['wind'].plot(
    ax=max_temp_axis,
    secondary_y=True,
    style='-s',
    color='steelblue',
    linewidth=1.5,
    alpha=0.6,
    xlim=('2012-01', '2012-03'),
    ylim=(0, 10),
    rot=45,
    label='Wind Speed'
)
wind_axis.set_ylabel('Wind (km/h)')
max_temp_axis.set_ylabel('Celsius (¬∞C)')
max_temp_axis.legend(['Max Temp'], loc='upper left')
max_temp_axis.right_ax.legend(['Wind Speed'], loc='upper right')
plt.show()
````
